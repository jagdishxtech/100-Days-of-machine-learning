```python
headers={"user-agent":"Mozilla/5.0 (iPhone; CPU iPhone OS 18_5 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/18.5 Mobile/15E148 Safari/604.1",
        "Accept-Language": "en-US,en;q=0.9" 
       }
company_name=[]
about=[]
rating=[] 
for i in range(1,10):
    url=f"https://www.ambitionbox.com/list-of-companies?page={i}"
    response=requests.get(url,headers=headers)
    webpage=response.text
    soup=BeautifulSoup(webpage, "lxml")
    
    for i in soup.find_all('h2',"companyCardWrapper__companyName"):
        company_name.append(i.text.strip())
    for i in (soup.find_all('span',"companyCardWrapper__interLinking")):
        about.append(i.text.strip())  
    for i in soup.select(".rating_text div"):
        rating.append(i.text.strip())
    time.sleep(2)    
d={"company":company_name,"about":about,"rating":rating}    
df=pd.DataFrame(d)
print("sucessfully run")
df
```
