```python
df=df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split((df.drop(columns="purchased")),df["purchased"],test_size=0.2)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
from sklearn.preprocessing import OrdinalEncoder
encode=OrdinalEncoder(categories=[["Poor","Average","Good"],["School","UG","PG"]])
encode.fit(X_test)
X_train_encode=encode.transform(X_train)
X_test_encode=encode.transform(X_test)
```

```python
X_train_encode=pd.DataFrame(X_train_encode,columns=X_train.columns)
X_test_encode=pd.DataFrame(X_test_encode,columns=X_test.columns)

```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(y_test)
y_train_encode=le.transform(y_train)
y_test_encode=le.transform(y_test)
y_test_encode
```
